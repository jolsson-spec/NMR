```python
# Simulates shot noise within a circuit
import numpy as np
import scipy as sp
import sympy
import matplotlib.pyplot as plt
from matplotlib.animation import *
import ffmpeg
```

```python
class shotnoise:
    
    # Simulates shot noise over one period of an AC circuit
    def currentconst(T, imax):
        # Constructs an expression for the current of a simple AC circuit given period T,
        # maximum current imax
        t = np.linspace(0, T, 1000)
        omega = 2*np.pi/T
        I = imax*np.cos(omega*T)
        return I, t
        
    def rmscurrent(Idc, T):
        # Calculates rms current given a period
        irms = np.sqrt(2*Idc*(1/T))
        return irms
        
    def randpoisson(irms, T, num):
        # Returns Poisson distributed samples of the current amplitude given an
        # rms current and time period
        var = irms/T
        k = irms
        return np.random.poisson(var, 1000)
        
    def schottky(irms, natural):
        # Returns Schottky's expression for the spectral density given an rms current
        if natural == True:
            sch = 2*irms
        else:
            sch = 2*sp.constants.e*irms
        return sch

    def randpoissoncont(lamb, k):
        return None
    
    
```
